```python
import os
import string



folder = '/content/drive/My Drive/Colab Notebooks/CLIGEN/2015parfiles/2015parfiles/2015parfiles'
outfile = '/content/drive/My Drive/Colab Notebooks/USCLIGEN_Annual_Precip.csv'
files = os.listdir( folder )

files = [f for f in files if f[:2] not in ['st', 'ak', 'hi', 'pi', 'pr', 'vi']]
files = [f for f in files if '.par' in f]
files = [os.path.join(folder, f) for f in files]
print(files)
ndays_months = [31, 28.25, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
with open(outfile, 'w') as fo:

  fo.write( 'stationID,annualP\n' )

  for f in files:
    print(f)
    top_skip_lines = 3

    all_lbl_chars = ((string.punctuation + string.ascii_letters)                \
                      .replace( '.', '' ))                                      \
                      .replace( '-', '' )

    with open(f) as fi:
      lines = fi.readlines()

    for i, line in enumerate(lines):

      for j, elem in enumerate(line):

        if elem in all_lbl_chars:
          label_end_i = j

        else:
          pass

      label = line[:label_end_i+1]
      label = label.strip()

      dataline = line[8:]
      row = [dataline[k*6:k*6+6] for k in range(12)]

      if label == 'MEAN P':
          meanp_row = [float(dataline[k*6:k*6+6]) for k in range(12)]
      elif label == 'P(W/W)':
          pww_row = [float(dataline[k*6:k*6+6]) for k in range(12)]
      elif label == 'P(W/D)':
          pwd_row = [float(dataline[k*6:k*6+6]) for k in range(12)]
      else:
          pass

    annualP = sum([ndays_months[i]*meanp_row[i]*(pwd_row[i]/(1 - pww_row[i] + pwd_row[i])) for i in range(12)])
    stationID = os.path.split(f)[-1][:-4]
    fo.write(stationID + ',' + str(annualP) + '\n')


```
