```
import os

folder = '/content/drive/My Drive/Colab Notebooks/CLIGEN/2015parfiles/2015parfiles/2015parfiles'
files = os.listdir(folder)
files = [f for f in files if f[:2] not in ['st', 'ak', 'hi', 'pi', 'pr', 'vi']]
files = [os.path.join(folder, f) for f in files if f not in ['ca049035.par']]
outfolder = '/content/drive/My Drive/GEE_Downloads'

area_files = []
for f in files:
  with open(f) as fi:
    lines = fi.readlines()
  row = lines[1].split( '=' )
  lat = float(row[1].split()[0])
  lon = float(row[2].split()[0])

  if lon >= -121. and lon <= -102. and lat <= 43. and lat >= 30.:
    area_files.append(f)

files = area_files


###############################################################################
'MEANP'
###############################################################################

data_dict = {}
for f in files:

  with open(f) as fi:
    lines = fi.readlines()

  xyline = lines[1]
  lat = float(xyline.split( 'LATT=' )[1].split( 'LONG=' )[0])
  lon = float(xyline.split( 'LONG=' )[1].split( 'YEARS=' )[0])
  xystr = str(lon) + ' ' + str(lat)

  line = lines[3]
  line_data = line[8:]
  data = [line_data[i*6:(i*6)+6].strip( '\n' ).lstrip().rstrip() for i in range(12)]
  data_dict[xystr] = data

for i in range(12):
  with open(os.path.join(outfolder, 'MEAN_' + str(i+1) + '.txt'), 'w') as fout:
    fout.write('x,y,data\n')
    for item in data_dict.items():
      x = item[0].split()[0]
      y = item[0].split()[1]
      data = item[1]
      fout.write( x + ',' + y + ',' + data[i] + '\n' )


###############################################################################
'SDEVP'
###############################################################################

data_dict = {}
for f in files:

  with open(f) as fi:
    lines = fi.readlines()

  xyline = lines[1]
  lat = float(xyline.split( 'LATT=' )[1].split( 'LONG=' )[0])
  lon = float(xyline.split( 'LONG=' )[1].split( 'YEARS=' )[0])
  xystr = str(lon) + ' ' + str(lat)

  line = lines[4]
  line_data = line[8:]
  data = [line_data[i*6:(i*6)+6].strip( '\n' ).lstrip().rstrip() for i in range(12)]
  data_dict[xystr] = data

for i in range(12):
  with open(os.path.join(outfolder, 'SDEV_' + str(i+1) + '.txt'), 'w') as fout:
    fout.write('x,y,data\n')
    for item in data_dict.items():
      x = item[0].split()[0]
      y = item[0].split()[1]
      data = item[1]
      fout.write( x + ',' + y + ',' + data[i] + '\n' )


###############################################################################
'SKEWP'
###############################################################################

data_dict = {}
for f in files:

  with open(f) as fi:
    lines = fi.readlines()

  xyline = lines[1]
  lat = float(xyline.split( 'LATT=' )[1].split( 'LONG=' )[0])
  lon = float(xyline.split( 'LONG=' )[1].split( 'YEARS=' )[0])
  xystr = str(lon) + ' ' + str(lat)

  line = lines[5]
  line_data = line[8:]
  data = [line_data[i*6:(i*6)+6].strip( '\n' ).lstrip().rstrip() for i in range(12)]
  data_dict[xystr] = data

for i in range(12):
  with open(os.path.join(outfolder, 'SKEW_' + str(i+1) + '.txt'), 'w') as fout:
    fout.write('x,y,data\n')
    for item in data_dict.items():
      x = item[0].split()[0]
      y = item[0].split()[1]
      data = item[1]
      fout.write( x + ',' + y + ',' + data[i] + '\n' )


###############################################################################
'MX5P'
###############################################################################

data_dict = {}
for f in files:

  with open(f) as fi:
    lines = fi.readlines()

  xyline = lines[1]
  lat = float(xyline.split( 'LATT=' )[1].split( 'LONG=' )[0])
  lon = float(xyline.split( 'LONG=' )[1].split( 'YEARS=' )[0])
  xystr = str(lon) + ' ' + str(lat)

  line = lines[14]
  line_data = line[8:]
  data = [line_data[i*6:(i*6)+6].strip( '\n' ).lstrip().rstrip() for i in range(12)]
  data_dict[xystr] = data

for i in range(12):
  with open(os.path.join(outfolder, 'MX5P_' + str(i+1) + '.txt'), 'w') as fout:
    fout.write('x,y,data\n')
    for item in data_dict.items():
      x = item[0].split()[0]
      y = item[0].split()[1]
      data = item[1]
      fout.write( x + ',' + y + ',' + data[i] + '\n' )


###############################################################################
'RATIO'
###############################################################################

data_dict = {}
for f in files:

  with open(f) as fi:
    lines = fi.readlines()

  xyline = lines[1]
  lat = float(xyline.split( 'LATT=' )[1].split( 'LONG=' )[0])
  lon = float(xyline.split( 'LONG=' )[1].split( 'YEARS=' )[0])
  xystr = str(lon) + ' ' + str(lat)

  pwwline = lines[6]
  pwdline = lines[7]
  pwwline_data = pwwline[8:]
  pwdline_data = pwdline[8:]

  pwwdata = [pwwline_data[i*6:(i*6)+6].strip( '\n' ).lstrip().rstrip() for i in range(12)]
  pwddata = [pwdline_data[i*6:(i*6)+6].strip( '\n' ).lstrip().rstrip() for i in range(12)]
  pwddata = [x if x != '.00' else '.01' for x in pwddata]

  data = [str(float(pwwdata[i])/float(pwddata[i])) for i, elem in enumerate(pwwdata)]
  data_dict[xystr] = data

for i in range(12):
  with open(os.path.join(outfolder, 'RATIO_' + str(i+1) + '.txt'), 'w') as fout:
    fout.write('x,y,data\n')
    for item in data_dict.items():
      x = item[0].split()[0]
      y = item[0].split()[1]
      data = item[1]
      fout.write( x + ',' + y + ',' + data[i] + '\n' )


###############################################################################
'TIMEPK'
###############################################################################

data_dict = {}
for f in files:

  with open(f) as fi:
    lines = fi.readlines()

  xyline = lines[1]
  lat = float(xyline.split( 'LATT=' )[1].split( 'LONG=' )[0])
  lon = float(xyline.split( 'LONG=' )[1].split( 'YEARS=' )[0])
  xystr = str(lon) + ' ' + str(lat)

  tpkline = lines[16]
  tpkline_data = tpkline[8:]
  tpkdata = [tpkline_data[i*6:(i*6)+6].strip( '\n' ).lstrip().rstrip() for i in range(12)]
  tpkdata = ['0' + x if x != '1.000' else '1.000' for x in tpkdata]
  data = tpkdata
  data_dict[xystr] = data


for i in range(12):
  with open(os.path.join(outfolder, 'TIMEPK_' + str(i+1) + '.txt'), 'w') as fout:
    fout.write('x,y,data\n')
    for item in data_dict.items():
      x = item[0].split()[0]
      y = item[0].split()[1]
      data = item[1]
      fout.write( x + ',' + y + ',' + data[i] + '\n' )


```
