```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy as scipy
import numpy as np
import os

baseFOLDER = '/content/drive/MyDrive/Colab Notebooks/Script Input Files'

precipEnsembleFILE = os.path.join(baseFOLDER, 'NEX_Ensemble_Stats_Precip_MEAN.csv')
tmaxEnsembleFILE = os.path.join(baseFOLDER, 'NEX_Ensemble_Stats_MaxTemp_MEAN.csv')


fig, ax = plt.subplots(figsize=(15, 5))

df_tseries = pd.read_csv(precipEnsembleFILE)
ccsm4_tseries = df_tseries['CCSM4']
canesm2_tseries = df_tseries['CanESM2']
miroc5_tseries = df_tseries['MIROC5']

df_ensemble = pd.read_csv(precipEnsembleFILE)
min_tseries = df_ensemble['min']
q25_tseries = df_ensemble['q25']
mean_tseries = df_ensemble['avg']
q75_tseries = df_ensemble['q75']
max_tseries = df_ensemble['max']

y = np.array(ccsm4_tseries)
x = np.array([i for i, elem in enumerate(range(0, 86))])
res = scipy.stats.linregress(x, y, alternative='greater')
slope = res.slope
print(res)

ax.plot(range(1985, 2071), min_tseries, linestyle='--', color='gray', linewidth=1, label='Ensemble Min/Max')
ax.plot(range(1985, 2071), mean_tseries, linestyle='--', color='black', linewidth=1, label='Ensemble Mean')
ax.fill_between(range(1985, 2071), q25_tseries, q75_tseries, linewidth=0, color='#272727', alpha=0.25, label='Ensemble Interquartile Range')
ax.plot(range(1985, 2071), max_tseries, linestyle='--', color='gray', linewidth=1)
ax.plot(range(1985, 2071), ccsm4_tseries, label='CCSM4')
ax.plot(range(1985, 2071), canesm2_tseries, label='CanESM2')
ax.plot(range(1985, 2071), miroc5_tseries, label='MIROC5')

ax.set_title('GCM Ensemble Statistics for Regional Annual Precip. (30-year Moving Avg.)', size=14)
ax.set_ylabel('Annual Precip. (mm)\n', size=12)
ax.set_xticks([1985, 2000, 2015, 2030, 2045, 2060, 2070])

# Shrink current axis's height by 15% on the bottom
box = ax.get_position()
ax.set_position([box.x0, box.y0 + box.height * 0.15,
                 box.width, box.height * 0.85])

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1),
          fancybox=True, shadow=True, ncol=6)

fig.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy as scipy
import numpy as np
import os

baseFOLDER = '/content/drive/MyDrive/Colab Notebooks/Script Input Files'
precipEnsembleFILE = os.path.join(baseFOLDER, 'NEX_Ensemble_Stats_Precip_MEAN.csv')
tmaxEnsembleFILE = os.path.join(baseFOLDER, 'NEX_Ensemble_Stats_MaxTemp_MEAN.csv')
tmaxEnsembleC6FILE = os.path.join(baseFOLDER, 'NEX_Ensemble_Stats_MaxTemp_MEAN_CMIP6.csv')
precipEnsembleC6FILE = os.path.join(baseFOLDER, 'NEX_Ensemble_Stats_Precip_MEAN_CMIP6.csv')
saveFILE = '/content/drive/MyDrive/GEE_Downloads/Fig8.jpg'

fig, ax = plt.subplots(2, 1, layout='constrained', figsize=(10, 4.5))
fig.subplots_adjust(hspace=20)

plt.rcParams['axes.grid'] = True

data_years = [40, 40, 34, 35, 33]

for ax_i in range(2):

  if ax_i == 0:

    df_tseries = pd.read_csv(precipEnsembleFILE)
    ccsm4_tseries = df_tseries['CCSM4']
    canesm2_tseries = df_tseries['CanESM2']
    miroc5_tseries = df_tseries['MIROC5']

    df_ensemble = pd.read_csv(precipEnsembleFILE)
    min_tseries = df_ensemble['min']
    q25_tseries = df_ensemble['q25']
    mean_tseries = df_ensemble['avg']
    q75_tseries = df_ensemble['q75']
    max_tseries = df_ensemble['max']

    df_ensemble = pd.read_csv(precipEnsembleC6FILE)
    meanc6_tseries = df_ensemble['avg']

    y = np.array(ccsm4_tseries)
    x = np.array([i for i, elem in enumerate(range(0, 86))])
    res = scipy.stats.linregress(x, y, alternative='greater')
    slope = res.slope
    print(res)
    ax[ax_i].text(1980, 423, '(a)', size=14)
    ax[ax_i].plot(range(1985, 2071), max_tseries, linestyle='--', color='gray', linewidth=1)
    ax[ax_i].plot(range(1985, 2071), mean_tseries, linestyle='--', color='black', linewidth=1, label='Ensemble Mean')
    ax[ax_i].plot(range(1985, 2071), ccsm4_tseries, label='CCSM4')
    ax[ax_i].plot(range(1985, 2071), min_tseries, linestyle='--', color='gray', linewidth=1, label='Ensemble Min/Max')
    ax[ax_i].plot(range(1985, 2071), miroc5_tseries, label='MIROC5')
    ax[ax_i].fill_between(range(1985, 2071), q25_tseries, q75_tseries, linewidth=0, color='#272727', alpha=0.25, label='Ensemble Interquartile Range')
    ax[ax_i].plot(range(1985, 2071), canesm2_tseries, label='CanESM2')
    ax[ax_i].plot(range(1985, 2071), meanc6_tseries, linestyle='--', color='purple', linewidth=1, label='Ensemble Mean CMIP6', zorder=-999)

    ax[ax_i].set_title('GCM Trends for Regional Annual Precip. (30-year Moving Avg.)', fontstyle='oblique', size=12)
    ax[ax_i].set_ylabel('Annual Precip. (mm)\n', size=10)
    ax[ax_i].set_xticks([1985, 2000, 2015, 2030, 2045, 2060, 2070])
    ax[ax_i].set_xticklabels([])

    # Shrink current axis's height by 15% on the bottom
    box = ax[ax_i].get_position()
    ax[ax_i].set_position([box.x0, box.y0 + box.height * 0.15,
                    box.width, box.height * 0.85])

  elif ax_i == 1:

    df_tseries = pd.read_csv(tmaxEnsembleFILE)
    ccsm4_tseries = (df_tseries['CCSM4'] - 32) * 5 / 9
    canesm2_tseries = (df_tseries['CanESM2'] - 32) * 5 / 9
    miroc5_tseries = (df_tseries['MIROC5'] - 32) * 5 / 9

    df_ensemble = pd.read_csv(tmaxEnsembleFILE)
    min_tseries = (df_ensemble['min'] - 32) * 5 / 9
    q25_tseries = (df_ensemble['q25'] - 32) * 5 / 9
    mean_tseries = (df_ensemble['avg'] - 32) * 5 / 9
    q75_tseries = (df_ensemble['q75'] - 32) * 5 / 9
    max_tseries = (df_ensemble['max'] - 32) * 5 / 9

    df_ensemble = pd.read_csv(tmaxEnsembleC6FILE)
    meanc6_tseries = (df_ensemble['avg'] - 32) * 5 / 9
    ax[ax_i].text(1980, 26, '(b)', size=14)
    ax[ax_i].plot(range(1985, 2071), max_tseries, linestyle='--', color='gray', linewidth=1)
    ax[ax_i].plot(range(1985, 2071), mean_tseries, linestyle='--', color='black', linewidth=1, label='Ensemble Mean')
    ax[ax_i].plot(range(1985, 2071), ccsm4_tseries, label='CCSM4')
    ax[ax_i].plot(range(1985, 2071), min_tseries, linestyle='--', color='gray', linewidth=1, label='Ensemble Min/Max')
    ax[ax_i].plot(range(1985, 2071), miroc5_tseries, label='MIROC5')
    ax[ax_i].fill_between(range(1985, 2071), q25_tseries, q75_tseries, linewidth=0, color='#272727', alpha=0.25, label='Ensemble Interquartile Range')
    ax[ax_i].plot(range(1985, 2071), canesm2_tseries, label='CanESM2')
    ax[ax_i].plot(range(1985, 2071), meanc6_tseries, linestyle='--', color='purple', linewidth=1, label='Ensemble Mean CMIP6', zorder=-999)

    ax[ax_i].set_title('GCM Trends for Regional Mean Daily High Temp. (30-year Moving Avg.)', fontstyle='oblique', size=12)
    ax[ax_i].set_ylabel('Mean Daily High Temp. (C)\n', size=10, labelpad=10)
    ax[ax_i].set_xticks([1985, 2000, 2015, 2030, 2045, 2060, 2070])

    # Shrink current axis's height by 15% on the bottom
    box = ax[ax_i].get_position()
    ax[ax_i].set_position([box.x0, box.y0 + box.height * 0.15,
                    box.width, box.height * 0.85])

    ax[ax_i].legend(loc='lower center', bbox_to_anchor=(0.5, -0.67),
              fancybox=True, shadow=True, ncols=4)

    ax[ax_i].grid(color='gray')

fig.savefig(saveFILE, dpi=600, bbox_inches='tight')


```
