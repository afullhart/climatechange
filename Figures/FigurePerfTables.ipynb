```python

import shutil
import os
import pandas as pd
import numpy as np

dataDIR = '/content/drive/My Drive/Colab Notebooks/FigureTables'

gcmLabels = ['CCSM4', 'CanESM2', 'MIROC5']
tableFiles = ['FOCAL_RMSE.csv', 'EBK_CV.csv', 'GB_CV.csv']
varLabels = ['MEANP 2000...2070 1...12', 'SDEVP 2000...2070 1...12', 'SKEWP 2000...2070 1...12']

data_rows = []
table = tableFiles[0]
for gcm in gcmLabels:
  fpath = os.path.join(dataDIR, gcm, table)
  with open(fpath) as f:
    lines = f.readlines()
  for i, var in enumerate(['mean', 'sdev', 'skew']):
    var_rows = []
    for line in lines:
      row = line.strip('\n').split(',')
      if var in line:
        var_rows.append(row)
    r_avg, p_avg, m_avg = 0.0, 0.0, 0.0
    for row in var_rows:
      r_avg += float(row[1])
      p_avg += float(row[2])
      m_avg += float(row[3])

    r_avg = round(r_avg/len(var_rows), 4)
    p_avg = round(p_avg/len(var_rows), 4)
    m_avg = round(m_avg/len(var_rows), 4)
    avgs = [varLabels[i], str(r_avg), str(p_avg), str(m_avg)]
    data_rows.append(avgs)

label, headr, gcms, rmse, pbias, mape = '| Parameter | ', '| ------ | ', '| GCM | ', '| RMSE | ', '| PBIAS | ', '| MAPE | '
for i, row in enumerate(data_rows):
  label += row[0] + ' | '
  headr += '------' + ' | '
  gcms += 'gcm' + ' | '
  rmse += row[1] + ' | '
  pbias += row[2] + ' | '
  mape += row[3] + ' | '

label = label[:-1]
headr = headr[:-1]
gcms = gcms[:-1]
rmse = rmse[:-1]
pbias = pbias[:-1]
mape = mape[:-1]
print(label)
print(headr)
print(gcms)
print(rmse)
print(pbias)
print(mape)
print('\n\n\n\n\n\n')




varLabels = ['MEANP 1974 1...12', 'SDEVP 1974 1...12', 'SKEWP 1974 1...12', 'RATIO 1974 1...12', 'TimePk 1974 1...12']

data_rows = []
table = tableFiles[1]
for gcm in gcmLabels:
  fpath = os.path.join(dataDIR, gcm, table)
  with open(fpath) as f:
    lines = f.readlines()
  for i, var in enumerate(['mean', 'sdev', 'skew', 'ratio', 'timepk']):
    var_rows = []
    for line in lines:
      row = line.strip('\n').split(',')
      if var in line:
        var_rows.append(row)
    r_avg, p_avg, m_avg = 0.0, 0.0, 0.0
    for row in var_rows:
      r_avg += float(row[1])
      p_avg += float(row[2])
      m_avg += float(row[3])

    r_avg = round(r_avg/len(var_rows), 4)
    p_avg = round(p_avg/len(var_rows), 4)
    m_avg = round(m_avg/len(var_rows), 4)
    avgs = [varLabels[i], str(r_avg), str(p_avg), str(m_avg)]
    data_rows.append(avgs)

label, headr, gcms, rmse, pbias, mape = '| Parameter | ', '| ------ | ', '| GCM | ', '| RMSE | ', '| PBIAS | ', '| MAPE | '
for i, row in enumerate(data_rows):
  label += row[0] + ' | '
  headr += '------' + ' | '
  gcms += 'gcm' + ' | '
  rmse += row[1] + ' | '
  pbias += row[2] + ' | '
  mape += row[3] + ' | '

label = label[:-1]
headr = headr[:-1]
gcms = gcms[:-1]
rmse = rmse[:-1]
pbias = pbias[:-1]
mape = mape[:-1]
print(label)
print(headr)
print(gcms)
print(rmse)
print(pbias)
print(mape)
print('\n\n\n\n\n\n')










varLabels = ['MX.5P 1974...2070 1...12']

data_rows = []
table = tableFiles[2]
for gcm in gcmLabels:
  fpath = os.path.join(dataDIR, gcm, table)
  with open(fpath) as f:
    lines = f.readlines()
  for i, var in enumerate(['mx5p']):
    var_rows = []
    for line in lines:
      row = line.strip('\n').split(',')
      if var in line:
        var_rows.append(row)
    r_avg, p_avg, m_avg = 0.0, 0.0, 0.0
    for row in var_rows:
      r_avg += float(row[1])
      p_avg += float(row[2])
      m_avg += float(row[3])

    r_avg = round(r_avg/len(var_rows), 4)
    p_avg = round(p_avg/len(var_rows), 4)
    m_avg = round(m_avg/len(var_rows), 4)
    avgs = [varLabels[i], str(r_avg), str(p_avg), str(m_avg)]
    data_rows.append(avgs)

label, headr, gcms, rmse, pbias, mape = '| Parameter | ', '| ------ | ', '| GCM | ', '| RMSE | ', '| PBIAS | ', '| MAPE | '
for i, row in enumerate(data_rows):
  label += row[0] + ' | '
  headr += '------' + ' | '
  gcms += 'gcm' + ' | '
  rmse += row[1] + ' | '
  pbias += row[2] + ' | '
  mape += row[3] + ' | '

label = label[:-1]
headr = headr[:-1]
gcms = gcms[:-1]
rmse = rmse[:-1]
pbias = pbias[:-1]
mape = mape[:-1]
print(label)
print(headr)
print(gcms)
print(rmse)
print(pbias)
print(mape)
print('\n\n\n\n\n\n')



```
